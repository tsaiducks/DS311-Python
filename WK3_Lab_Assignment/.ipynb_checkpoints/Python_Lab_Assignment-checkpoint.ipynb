#### Find the passing rate for math and reading (above 70 points)

```python
# Find the total counts of math result
math_count = school_data_complete['math_score'].count()

# Find the counts for math result in each school that pass 70 or higher
pass_math = school_data_complete[school_data_complete['math_score'] >= 70].groupby('school_name')['math_score'].count()

pass_math

# Calculate the math passing rate
math_pass_rate = ((pass_math / math_count) * 100).round(2)

math_pass_rate


```

```python
# Find the total counts of read result
reading_count = school_data_complete['reading_score'].count()

# Find the counts for read result in each school that pass 70 or higher
pass_reading = school_data_complete[school_data_complete['reading_score'] >= 70].groupby('school_name')['reading_score'].count()

# Calculate the read passing rate
reading_pass_rate = ((pass_reading / reading_count) * 100).round(2)

reading_pass_rate
```

```python
# Calculate the overall passing rate (average of the math and reading passing rate)
overall_pass_rate = (math_pass_rate + reading_pass_rate) / 2

overall_pass_rate
```

## Scores by School Spending

* Create a table that breaks down school performances based on average Spending Ranges (Per Student). Use 4 reasonable bins to group school spending. Include in the table each of the following:
  * Average Math Score
  * Average Reading Score
  * % Passing Math
  * % Passing Reading
  * Overall Passing Rate (Average of the above two)

```python
# Sample bins. Feel free to create your own bins.
spending_bins = [0, 585, 615, 645, 675]
group_names = ["<$585", "$585-615", "$615-645", "$645-675"]
```

```python
# Create a new column to show budget per student in each row
school_data_complete['Budget per Student'] = (school_data_complete['budget'] / school_data_complete['size']).round(2)

school_data_complete.head(6)
```

```python
# Create a new column to define the spending ranges per student
school_data_complete['Spending Ranges Per Student'] = pd.cut(school_data_complete['Budget per Student'], spending_bins, labels = group_names)

school_data_complete.head(6)
```

```python
# Calculate the average math score within each spending range
avg_math_score_spending = school_data_complete.groupby('Spending Ranges Per Student')['math_score'].mean().round(2)

avg_math_score_spending
```

```python
# Calculate the percentage passing rate for math in each spending range
pass_math_spending_rate = (school_data_complete[school_data_complete['math_score'] >= 70].groupby('Spending Ranges Per Student')['math_score'].count()\
                           / school_data_complete.groupby('Spending Ranges Per Student')['math_score'].count() * 100).round(2) 

pass_math_spending_rate
```

```python
# Calculate the percentage passing rate for reading in each spending range

pass_reading_spending_rate = (school_data_complete[school_data_complete['reading_score'] >= 70].groupby('Spending Ranges Per Student')['reading_score'].count()\
                           / school_data_complete.groupby('Spending Ranges Per Student')['reading_score'].count() * 100).round(2) 

pass_reading_spending_rate
```

```python
# Calculate the percentage overall passing rate in each spending range
pass_overall_spending_rate = ((pass_math_spending_rate + pass_reading_spending_rate) / 2).round(2)

pass_overall_spending_rate
```
